```python
data['review'] = data['review'].str.lower()

data.head()
```

```python
# Apply stemming technique on review data 
```

```python
ps = PorterStemmer()

data['review'] = data['review'].apply(
    lambda x: " ".join([ps.stem(word) for word in x.split()])
)

data.head()
```

```python
# Apply Bag of Word technique and convert into dataframe 
```

```python
cv = CountVectorizer(max_features=20)

X = cv.fit_transform(data['review'])

bow_df = pd.DataFrame(X.toarray(), columns=cv.get_feature_names_out())

bow_df.head()
```

```python
# Difference between stemming and lemmatization ? 
```

```python
# | Stemming                                   | Lemmatization                              |
# | Removes suffix from words to get root form | Converts word to its dictionary base form  |
# | Uses simple rules                          | Uses vocabulary and morphological analysis |
# | Faster but less accurate                   | Slower but more accurate                   |

# Example:

# Stemming:
# playing → play
# studies → studi

# Lemmatization:
# playing → play
# studies → study
```

```python
# The NLP pipeline is a sequence of steps used to process and analyze text data.

# Steps in NLP pipeline:

# Text Collection – gather text data

# Text Cleaning – remove punctuation and special characters

# Tokenization – split text into words

# Stopword Removal – remove common words like is, the

# Stemming/Lemmatization – convert words to root form

# Feature Extraction – Bag of Words / TF-IDF

# Model Building – classification or prediction
```
